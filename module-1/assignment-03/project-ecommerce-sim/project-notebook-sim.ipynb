## preprocessing: Changing rating of products to float, review_count to int and price to float

```python
import re

products=[(df_dslr,'dslr'),
(df_keyboard,'keyboard'),
(df_monitor,'monitor'),
(df_mouse,'mouse'),
(df_processor,'processor'),
(df_smartphone,'smartphone')]

for product in products:
    product[0]['rating']=product[0]['rating'].apply(lambda x: round(float(df_smartphone.at[2,'rating'].split(' out')[0]),2))
    product[0]['review_count']=product[0]['review_count'].apply(lambda x: int(''.join(x.split(','))))
    product[0]['price']=product[0]['price'].apply(lambda x: round(float(re.sub('[^1-9.]','',x)),2))
    # product[0].to_csv(f'assignment-03/project-ecommerce/datasets/ecom-sales_data/ecom-sales{product[1]}_processed.csv')

```

```python
grouped=df_sales.groupby('product_type')
```

```python
# fig,axes=plt.subplots(nrows=df_sales['prdct_type'].nunique(), ncols=1)
for i, group in enumerate(list(grouped.groups.keys())):
    fig, ax = plt.subplots(ncols=1,nrows=1,figsize=(8,4))
    local_df=grouped.get_group(group).groupby('title').agg({'total_sale':'sum'}).sort_values('total_sale',ascending=False).head(3)
    x=list(map(lambda x: x[:40],local_df.index))
    y=local_df['total_sale'].values
    y_labels=[]
    [''.join([str(i)[::-1]+',' if (t+1)%3==0 else str(i.round(2))[::-1][t] for t in range(len(str(i.round(2))))])[::-1]+'$' for i in y]
    for t in y:
        if t-int(i)>0:
            i_str=str(t.round(2))
            nat_part=i_str[:-3][::-1] if round(t,2)-round(t,1)>0 else i_str[:-2][::-1]
            dec_part=i_str[-3:] if round(t,2)-round(t,1)>0 else i_str[-2:]
            y_labels.append(''.join([nat_part[q]+',' if (q+1)%3==0 else nat_part[q] for q in range(len(nat_part))])[::-1]+dec_part)
        else:
            t=str(int(t))
            nat_part=t[::-1]
            y_labels.append(''.join([nat_part[q]+',' if (q+1)%3==0 else nat_part[q] for q in range(len(nat_part))])[::-1])

    plt_legend= list(map(lambda x: 'ASIN: ' + df_sales[df_sales['title']==x]['ASIN'].unique()[0]+'\n',local_df.index))
    plt_legend=[plt_legend[i]+'\n'+'Total sale: '+y_labels[i] +'\n' for i in range(len(plt_legend))]
    colors=['darkorchid','mediumorchid','orchid']
    colors_dict={x[i]:colors[i] for i in range(len(colors))}
    ax.bar(x=x,height=y,width=0.4,color=colors)
    handles=[plt.Rectangle((0,0),1,1, color=colors_dict[i]) for i in x]
    ax.set_xticklabels(x, rotation=350, horizontalalignment='left',fontsize='xx-large')
    ax.set_xlabel('Product',fontsize='xx-large')
    ax.set_ylabel('Total Sales in $US ',fontsize='xx-large')
    ax.set_yticks(y)
    ax.set_yticklabels([round(t,2) for t in y],rotation=30,fontsize='xx-large') if group not in ['keyboard','monitor'] else ax.set_yticklabels([])
    ax.grid(b=True,axis='y')
    ax.set_title('Top selling items in segment: '+group.capitalize(),fontsize='xx-large')
    ax.legend(handles,plt_legend,loc='upper right',bbox_to_anchor=(1, 0.04, 0.55, 1),fontsize='x-large')
    # print(plt_legend)
    plt.savefig(f'plots/top_selling_{group}',bbox_inches='tight')
```
