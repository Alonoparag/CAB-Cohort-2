* This Notebook is used to merge the individual csv files, representing monthly data from capital bikeshare, into a single dataframe.
* I chose to use notebook because it was easier to diagnose errors on the fly than using a python file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import seaborn as sns
```

```python
df_05=pd.read_csv('../../datasets/cbs-historical/202005-capitalbikeshare-tripdata.csv')
df_06=pd.read_csv('../../datasets/cbs-historical/202006-capitalbikeshare-tripdata.csv')
df_07=pd.read_csv('../../datasets/cbs-historical/202007-capitalbikeshare-tripdata.csv')
df_08=pd.read_csv('../../datasets/cbs-historical/202008-capitalbikeshare-tripdata.csv')
df_09=pd.read_csv('../../datasets/cbs-historical/202009-capitalbikeshare-tripdata.csv')
df_10=pd.read_csv('../../datasets/cbs-historical/202010-capitalbikeshare-tripdata.csv')
df_11=pd.read_csv('../../datasets/cbs-historical/202011-capitalbikeshare-tripdata.csv')
df_12=pd.read_csv('../../datasets/cbs-historical/202012-capitalbikeshare-tripdata.csv')
df_21_1=pd.read_csv('../../datasets/cbs-historical/202101-capitalbikeshare-tripdata.csv')
df_21_2=pd.read_csv('../../datasets/cbs-historical/202102-capitalbikeshare-tripdata.csv')
df_21_3=pd.read_csv('../../datasets/cbs-historical/202103-capitalbikeshare-tripdata.csv')
df_21_4=pd.read_csv('../../datasets/cbs-historical/202104-capitalbikeshare-tripdata.csv')

display(df_05.head(1))
display(df_06.head(1))
display(df_07.head(1))
display(df_08.head(1))
display(df_09.head(1))
display(df_10.head(1))
display(df_11.head(1))
display(df_12.head(1))
display(df_21_1.head(1))
display(df_21_2.head(1))
display(df_21_3.head(1))
display(df_21_4.head(1))
```

```python
df = pd.concat([df_05,
df_06,
df_07,
df_08,
df_09,
df_10,
df_11,
df_12,
df_21_1,
df_21_2,
df_21_3,
df_21_4])

df.shape
```

```python
df.started_at=df.started_at.apply(lambda x: np.datetime64(x))
df.ended_at=df.ended_at.apply(lambda x: np.datetime64(x))

df.sort_values('started_at',inplace = True)
df.reset_index(inplace = True, drop = True)

df.to_csv('../../datasets/bikeshare-final.csv')
```
